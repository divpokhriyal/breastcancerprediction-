```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset

bcancer_dataset = pd.read_csv('BreastCancerDataset.csv')
```

```python
bcancer_dataset
```

```python
#label encoding using Pandas - replacing M (Malignant) 0 with B(Benign) and with 1
bcancer_dataset.replace({"diagnosis":{'M':0, 'B':1}}, inplace = True)
```

```python
bcancer_dataset
```

```python
#Dropping the last column -- NaN values and id

bcancer_dataset = bcancer_dataset.drop(['Unnamed: 32', 'id'], axis=1)
```

```python
bcancer_dataset
```

```python
bcancer_dataset.info()

# no missing values 
```

```python
#statistical measures about the dataset

bcancer_dataset.describe()
```

```python
#checking the distribution of the target variable - diagnosis

bcancer_dataset['diagnosis'].value_counts()
```

```python
bcancer_dataset.groupby('diagnosis').mean()
```

```python
#seperating the features and target value

X = bcancer_dataset.drop(columns='diagnosis', axis =1)
Y = bcancer_dataset['diagnosis']
```

```python
X
```

```python
Y
```

```python
#Splitting data into training and testing data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Model Training - Logistic Regression 

model =LogisticRegression()
```

```python
#Training the Logistic Regression Model using Training data

model.fit(X_train, Y_train)
```

### Model Evaluation using accuracy_score

```python
#Accuracy of training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)

print('Accuracy of training data =', training_data_accuracy)
```

```python
#Accuracy of testing data 

X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)

print('Accuracy of testing data =', testing_data_accuracy)
```

### Building a predictive system 

```python
input_data = (18.25, 19.98, 119.6, 1040, 0.09463, 0.109, 0.1127, 0.074, 0.1794, 0.05742, 0.4467, 0.7732, 3.18, 53.91, 0.004314, 0.01382, 0.02254, 0.01039, 0.01369, 0.002179, 22.88, 27.66, 153.2, 1606, 0.1442, 0.257, 0.3784, 0.1932, 0.3063, 0.08368
)
```

```python
#changing input_data into a numpy array

input_data_array = np.asarray(input_data)

#reshaping the numpy array as predicting for only one datapoint

input_data_reshaped = input_data_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)

if (prediction[0]==0):
    print('Malignant')
    
else:
    print('Benign')

print(prediction)
```
